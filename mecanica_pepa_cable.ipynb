```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(0, 5, 10000)
r_0 = 0.2
w = 1
phi = w*t

r = r_0*np.cosh(phi)

fig, (ax1, ax2) = plt.subplots(1, 2, subplot_kw={'projection': 'polar'})

ax1.plot(phi, r)

ax1.set_rlabel_position(67.5)

ax1.set_title(r'$r(\phi)$ vs $\phi$ con $\omega$ = 1 rad/s', va='bottom')

ax2.plot(10*phi, r_0*np.cosh(10*phi), color='r')
ax2.set_title(r'$r(\phi)$ vs $\phi$ con $\omega$ = 10 rad/s', va='bottom')
ax2.set_rlabel_position(67.5)


fig.tight_layout()

fig.savefig('orbita')
```

```python
from turtle import right


t = np.linspace(0, 5, 10000)
r_0 = 0.2
w = 1

r = r_0*np.cosh(w*t)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)

ax1.plot(t, r)
ax1.set_title(r'$r(t)$ vs $t$ con $\omega$ = 1 rad/s', va='bottom')
ax1.set_ylabel(r'$r(t)$ (unidades)')
ax1.set_xlabel(r't(s)')
ax1.set_xticks([1,2,3,4,5])

ax2.plot(t, r_0*np.cosh(10*t), color='r')
ax2.set_title(r'$r(t)$ vs $t$ con $\omega$ = 10 rad/s', va='bottom')
ax2.set_ylabel(r'$r(t)$ (unidades)')
ax2.set_xlabel(r't(s)')
ax2.set_xticks([1,2,3,4,5])


ax3.plot(t, (180/np.pi)*w*t)
ax3.set_title(r'$\phi$(t) vs $t$ con $\omega$ = 1 rad/s', va='bottom')
ax3.set_ylabel(r'$\phi(t)$ (grados)')
ax3.set_xlabel(r't(s)')
ax3.set_xticks([1,2,3,4,5])


ax4.plot(t, (180/np.pi)*10*t, color='r')
ax4.set_title(r'$\phi$(t) vs $t$ con $\omega$ = 10 rad/s', va='bottom')
ax4.set_ylabel(r'$\phi(t)$ (grados)')
ax4.set_xlabel(r't(s)')
ax4.set_xticks([1,2,3,4,5])


fig.tight_layout()

fig.savefig('parametricas')
```
